[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def time_check(func):
    def wrapper(*args,**kwargs):
        st = time.time()
        rt = func(*args,**kwargs)
        et =time.time()
        print(f"함수 실행 시간:{et-st:.2f}초")
        return rt
    return wrapper

@time_check
def hello():
    time.sleep(3)
    print("안녕하세요 반갑습니다.")


hello()
````
